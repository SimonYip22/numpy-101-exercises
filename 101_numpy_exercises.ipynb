```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
sepallength = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0])

def softmax(x):
    """
    Compute softmax values for each sets of scores in x.

    Softmax is a function that takes as input a vector of real numbers and normalizes it into a probability distribution. 
    The output values are in the range (0, 1) and sum to 1. 

    The softmax function is often used in machine learning, particularly in classification problems, to convert raw scores (logits) into probabilities.
    """
    e_x = np.exp(x - np.max(x)) # subtracting np.max(x) for numerical stability
    return e_x / e_x.sum(axis=0) # divide the exponentials by the sum of exponentials to get probabilities

print(softmax(sepallength))
```

## 47. Clip Array Values to a Range

From the array a, replace all values greater than 30 with 30 and all values less than 10 with 10.

Difficulty Level: L2

Desired Output:

[ 27.63  14.64  21.8   30.    10.    10.    30.    30.    10.    29.18  30.
  11.25  10.08  10.    11.77  30.    30.    10.    30.    14.43]
  
Why this matters: Clipping values to a valid range is essential for outlier handling and ensuring inputs stay within expected bounds for models.

```python
# same random numbers 
np.random.seed(100) 

# generate 20 random numbers between 1 and 50 from a uniform distribution
a = np.random.uniform(1, 50, 20)

# Solution 1: Using np.clip
print(np.clip(a, a_min=10, a_max=30)) # force all values into a fixed range between 10 and 30.

# Solution 2: Using np.where
# nested conditional
# if_true: if a < 10, return 10
# if_false: if a > 30, return 30 if_true, if_false return a
print(np.where(a < 10, 10, np.where(a > 30, 30, a))) 

# Solution 3: Using boolean indexing
result = a.copy()
result[result > 30] = 30
result[result < 10] = 10

print(result)
```

## 48. Get the Positions of Top N Values

Get the positions of the top 5 maximum values in the array a.

Difficulty Level: L2

Desired Output:

[18 7 3 10 15]

Why this matters: Finding top-N indices is used in ranking systems, recommendation engines, and selecting the highest-confidence predictions.

```python
np.random.seed(100)
a = np.random.uniform(1, 50, 20)

# Solution 1:
# argsort() returns the indices that would sort an array.
index_sorted = np.argsort(a)

# Get the last 5 elements of the sorted array, which are the top 5 largest values.
top_5 = index_sorted[-5:]
print(top_5)

# Get the actual values
print(a[top_5])

# Solution 2:
# Simpler version of solution 1
print(a.argsort()[-5:])

# Solution 3:
# Does partial sorting
# argpartition() returns the indices that would partition an array into two parts: 
# the first part contains the k smallest elements, and the second part contains the rest of the elements.
# 5 means the 5 smallest values of -a are in the first 5 positions, the rest are left.
#[:5] selects the first 5 indices, they will be the 5 largest values but won't be sorted.
print(np.argpartition(-a, 5)[:5])

# Get the actual values:
print(a[a.argsort()][-5:])
```
